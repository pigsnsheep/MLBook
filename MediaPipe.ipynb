```python
import math
def calculate_angle(a,b,c):
    a = np.array(a) #Shoulder
    b = np.array(b) #Elbow
    c = np.array(c) #Wrist

    angle = abs(np.arctan2(c[1]-b[1], c[0]-b[0]) - np.arctan2(a[1]-b[1], a[0]-b[0]))*180/math.pi    
  

    if angle > 180:
        angle = 360-angle
    
    return angle

counter = 0


cap = cv2.VideoCapture(0)
with mp_pose.Pose(min_detection_confidence=0.5, min_tracking_confidence=0.5) as pose:
    while cap.isOpened():
        ret, frame = cap.read()

        #Mediapipe requires RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False

        #Detection
        results = pose.process(image)

        #Color it back
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        try:
            landmarks = results.pose_landmarks.landmark
            print("hi")

 

            left_shoulder = np.array([landmarks[11].x,landmarks[11].y])
            left_elbow = np.array([landmarks[13].x,landmarks[13].y])
            left_wrist = np.array([landmarks[15].x,landmarks[15].y])

            image = cv2.putText(image, str(counter), (50,200), cv2.FONT_HERSHEY_SIMPLEX,  
                   10, (255,255,255), 5, cv2.LINE_AA) 

            print(calculate_angle(left_shoulder, left_elbow, left_wrist))
        except:
            pass

        mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_pose.POSE_CONNECTIONS)


        cv2.imshow("Test", image)


        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
    cap.release()
    cv2.destroyAllWindows()
```
